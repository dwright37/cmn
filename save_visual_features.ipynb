# Get ROI pool activations for all of the validation images

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

from util.refgoog_baseline_train.roi_data_reader import DataReader
from models import fastrcnn_vgg_net

%matplotlib inline

outdir = '/hdd/dustin/data/gref_train_1000_roifwdprop'
```

```python
def getFileBaseName(filename):
    name = os.path.basename(filename)
    return name[:name.rfind('.')]
```

```python
imdb_file = './dwright-exp-image-ranking/imdb_train_1000.json'
vocab_file = './word_embedding/vocabulary_72700.txt'
im_mean = fastrcnn_vgg_net.channel_mean
reader = DataReader(imdb_file, vocab_file, im_mean, shuffle=True)
```

```python
im_batch = tf.placeholder(tf.float32, [1, None, None, 3]) 
bbox_batch = tf.placeholder(tf.float32, [None, 5]) 

vis_feature = fastrcnn_vgg_net.vgg_roi_fc7(im_batch, bbox_batch, "vgg_local", apply_dropout=False)
```

```python
# Start Session
sess = tf.Session(config=tf.ConfigProto(gpu_options=tf.GPUOptions(allow_growth=True)))

# Snapshot saver
snapshot_saver = tf.train.Saver()
snapshot_saver.restore(sess, './downloaded_models/vgg_pickle.tfmodel')
```

```python
total_size = []

for b in range(reader.num_batch):
    batch = reader.read_batch()

    #Get output path name
    filename = getFileBaseName(batch['im_path'])
    outfile = outdir + '/' + filename + '.npz'
    
    feat = sess.run(vis_feature, feed_dict={im_batch: batch['im_batch'], bbox_batch: batch['bbox_batch']})
    
    with open(outfile, 'wb') as f:
        np.savez(f, vis_feat=feat, spatial_feature=batch['spatial_batch'], image_path=batch['im_path'])

    total_size.append(os.path.getsize(batch['im_path']) / 1024.)
    
    print(str(sum(total_size)) + " KB")
    
```
